# visualizing n-dimensional data with contour plots

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
X,Y = np.meshgrid(x,y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.contour(X,Y,Z,levels=10)
plt.colorbar()

plt.xlabel('X')
plt.ylabel('Y')

plt.title('Contour plot')
plt.show()
```

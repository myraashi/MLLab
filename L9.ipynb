# Agglomerative clustering based on complete linkage and single linkage

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

data = np.array([
    [9,4],
    [25,8],
    [24,8],
    [36,12]
])
```

```python
def singlelinkage(data):
    n=data.shape[0]
    proximitymatrix = np.zeros((n,n))
    
    for i in range(n):
        for j in range(i+1,n):
            proximitymatrix[i][j] = np.min(np.linalg.norm(data[i]-data[j]))
            proximitymatrix[j][i] = proximitymatrix[i][j]
    return proximitymatrix
def completelinkage(data):
    n=data.shape[0]
    proximitymatrix = np.zeros((n,n))
    
    for i in range(n):
        for j in range(i+1,n):
            proximitymatrix[i][j] = np.max(np.linalg.norm(data[i]-data[j]))
            proximitymatrix[j][i] = proximitymatrix[i][j]
    return proximitymatrix
```

```python
smatrix = singlelinkage(data)
print("Single linkage matrix:\n",smatrix)
cmatrix = completelinkage(data)
print("Complete linkage matrix:\n",cmatrix)
```

```python
linkage_matrix = linkage(data,method='single')
plt.figure(figsize=(10,5))
dendrogram(linkage_matrix)
plt.xlabel('Data points')
plt.ylabel('Distance')
plt.title('Single linkage')
plt.show()
```

```python
lmatrix = linkage(data,method='complete')
plt.figure(figsize=(10,5))
dendrogram(lmatrix)
plt.xlabel('Data points')
plt.ylabel('Distance')
plt.title('Complete linkage')
plt.show()
```
